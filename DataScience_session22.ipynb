```python
# %matplotlib inline
# import matplotlib.pyplot as plt
from scipy import misc
from sklearn import cluster

# racoon = misc.face(gray = True)
racoon = misc.face()
from skimage import io
io.imshow(racoon)
print("beofore reshape")
io.show()
```

Examine the dimensions of the image

```python
racoon.shape
```

```python
x, y,z = racoon.shape
image_2d = racoon.reshape(x*y, z)
image_2d.shape
```

Create clusters

```python
kmeans_cluster = cluster.KMeans(n_clusters=5)
kmeans_cluster.fit(image_2d)
cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
```

```python
newimage = cluster_centers[cluster_labels].reshape(x, y,z)
```

```python
io.imshow(newimage)
print("After Clustering")
io.show()
```
